# Avocado prices & sales

In this notebook, we're going to explore historical data on avocado prices and sales volume in multiple US markets. We will do the following tasks: 

* Implement a simple program to read a csv document and print the content
* Create an class to represent the Avocado prices (i.e. date, AveragePrice, TotalVolume, TotalBags, SmallBags, LargeBags, XLargeBags,Year, Region)
* Create an object for each row of the given csv document and put all of them into a list
* Print the contents of the list in a for loop in two different ways:
    i) by using the print() function; 
    ii)by using printer() function implemented by you in the Avocado class
    



The following code snippet will do the first task. It will read every single line of the csv document, extract the values by using column names, and print them in a for loop.

```python
import functools
```

```python
class Avocado:
    def __init__(self, AveragePrice, TotalVolume, TotalBags, SmallBags, LargeBags, XLargeBags, Year, Region):
        self.contents = [AveragePrice, TotalVolume, TotalBags, SmallBags, LargeBags, XLargeBags, Year, Region]
        self.labels =   ["AveragePrice", "TotalVolume", "TotalBags", "SmallBags", "LargeBags", "XLargeBags", "Year", "Region"]
    def printer(self):
        combilist = list(map(lambda x,y: str(x) + ": " + str(y), self.labels, self.contents))
        print(str(" | ".join(combilist)))
    def __str__(self):
        combilist = list(map(lambda x,y: str(x) + ": " + str(y), self.labels, self.contents))
        return " | ".join(combilist)
```

```python
av = Avocado(1,2,3,4,5,6,7,8)
av.printer()
print(av)
```

```python
import csv
from datetime import datetime
import functools


if __name__ == '__main__':

    with open('avocado.csv') as csvfile:
        readCSV = csv.DictReader(csvfile, delimiter=',')
        avocadosales = []
        for col in readCSV:
            avocadosales.append(Avocado(col['AveragePrice'],
                                        col['Total Volume'],
                                        col['Total Bags'],
                                        col['Small Bags'],
                                        col['Large Bags'],
                                        col['XLarge Bags'],
                                        col['year'],
                                        col['region']))

    for a in avocadosales:
      a.printer()
      print(a)
```
